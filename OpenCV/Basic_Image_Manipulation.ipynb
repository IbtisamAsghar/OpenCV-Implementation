# Cropping images

```python
img_Nz_bgr = cv2.imread('New_Zealand_Boat.jpg',1)
img_NZ_RGB = img_Nz_bgr[:,:,::-1]
plt.imshow(img_NZ_RGB)
plt.show()
```

# Crop Out the middle region of the image


```python
cropped_img = img_NZ_RGB[200:400,300:600]
plt.imshow(cropped_img)
plt.show()
```

# Resizing Images

Method: 1 Specifying scaling factor using fx and fy

```python
resize_img = cv2.resize(cropped_img,None,fx=3,fy=3)
plt.imshow(resize_img)
plt.show()
```

Method: 2 Specifying exact size of the output image

```python
desired_width = 100
desired_height = 200

dim = (desired_width , desired_height)

resize_croppped_img = cv2.resize(cropped_img , dsize=dim , interpolation=cv2.INTER_AREA)

plt.imshow(resize_croppped_img )
plt.show()
```

```python
cropped_img.shape[1] , cropped_img.shape[0]
```

Resize while maintaing aspect ration

```python
desired_width = 100
aspect_ratio = desired_width / cropped_img.shape[1]
desired_height = int(cropped_img.shape[0] * aspect_ratio)

dim = (desired_width , desired_height)

cropped_img_resize_2 = cv2.resize(cropped_img , dsize=dim , interpolation=cv2.INTER_AREA)
plt.imshow(cropped_img_resize_2)
plt.show()
```

```python
# Swap channel order
resize_img = resize_img [:, :, ::-1]

# Save resized image to disk
cv2.imwrite("resize_imged.png", resize_img)

# Display the cropped and resized image
Image(filename="resize_imged.png")
```

# Flipping the Images

```python
img_Nz_rgb_flipped_horz = cv2.flip(img_NZ_RGB,1) # flips the image horizontally
img_Nz_rgb_flipped_vert = cv2.flip(img_NZ_RGB,0) # flips the image vertically 
img_Nz_rgb_flipped_both = cv2.flip(img_NZ_RGB,-1) # flipped to 180 degrees

# Displaying the images
plt.figure(figsize=(18,5))
plt.subplot(141);plt.imshow(img_Nz_rgb_flipped_horz);plt.title("Horizontal Flip");
plt.subplot(142);plt.imshow(img_Nz_rgb_flipped_vert);plt.title("Horizontal Flip");
plt.subplot(143);plt.imshow(img_Nz_rgb_flipped_horz);plt.title("Horizontal Flip");
```
